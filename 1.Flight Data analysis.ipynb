```python
# if i want to seperate only date / day

df['Date']=df['Date_of_Journey'].str.split('/').str[0]   #---FOR DAY
df['Month']=df['Date_of_Journey'].str.split('/').str[1]   #---FOR MONTH
df['Year']=df['Date_of_Journey'].str.split('/').str[2]   #---FOR YEAR
```

```python
# to convert data type of column

df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```
